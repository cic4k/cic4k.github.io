```python
#ouvrir fichier
def lire_fichier(chemin):
    f = open(chemin, encoding="utf-8")
    chaine = f.read()
    f.close()
    return chaine

toto = lire_fichier('13846-0.txt')
titi = lire_fichier('4300-0.txt')
tutu = lire_fichier('pg44217.txt')
#decoupe en mots
def decouper_mots(chaine):
    liste_mots = chaine.split()
    return liste_mots
mots_toto = decouper_mots(toto)
mots_titi = decouper_mots(titi)
mots_tutu = decouper_mots(tutu)
#dico longueurs

def get_effectifs(liste_mots):
    dic = {}
    for mot in liste_mots:
        longueur = len(mot)
        if longueur not in dic:
            dic[longueur]=1
        else:
            dic[longueur] = dic[longueur]+1
    return dic

dic_toto = get_effectifs(mots_toto)
dic_titi = get_effectifs(mots_titi)
dic_tutu = get_effectifs(mots_tutu)

def dic2liste(dic):
    liste_effectifs = []
    for longueur in range(1,30):
        if longueur in dic:
            liste_effectifs.append(dic[longueur])
        else:
            liste_effectifs.append(0)
    return liste_effectifs
liste_effectifs_toto = dic2liste(dic_toto)
liste_effectifs_titi = dic2liste(dic_titi)
liste_effectifs_tutu = dic2liste(dic_tutu)

#range les effectifs pour chaque longueur
#graphique
```

```python
#ouvrir fichier
def lire_fichier(chemin):
    f = open(chemin, encoding="utf-8")
    chaine = f.read()
    f.close()
    return chaine
def decouper_mots(chaine):
    liste_mots = chaine.split()
    return liste_mots
def get_effectifs(liste_mots):
    dic = {}
    for mot in liste_mots:
        longueur = len(mot)
        if longueur not in dic:
            dic[longueur]=1
        else:
            dic[longueur] = dic[longueur]+1
    return dic
def dic2liste(dic):
    liste_effectifs = []
    for longueur in range(1,30):
        if longueur in dic:
            liste_effectifs.append(dic[longueur])
        else:
            liste_effectifs.append(0)
    return liste_effectifs

#range les effectifs pour chaque longueur
#graphique
```

```python
import matplotlib.pyplot as pyplot

liste_textes = ['13846-0.txt', '4300-0.txt', 'pg44217.txt', 'fi.txt']
for chemin in liste_textes:
    #étape 1 :lire fichier
    toto= lire_fichier(chemin)
    #étape 2: découper en mots
    mots_toto = decouper_mots(toto)
    #étape 3: calculer les effectifs
    dic_toto = get_effectifs(mots_toto)
    liste_effectifs_toto = dic2liste(dic_toto)
    #tracer une courbe
    pyplot.plot(liste_effectifs_toto)

pyplot.show()
```
